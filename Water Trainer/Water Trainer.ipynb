# Water Pokemon:
## Water pokemon on found in and around water. One thing that Pittsburgh has is an abundance of rivers, streams, and tributaries. Water Pokemon are also gainfully employed at water parks and public features. This is the data we will use to see where we can find some water pokemon around the city.   

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
waterfeatures_df = pd.read_csv("Pittsburgh Water Features.csv")
```

## We can now create a new DataFrame to show us which neighborhoods have the most water features. 

```python
WF_count = waterfeatures_df.groupby("neighborhood")['name'].count().reset_index()
WF_count.columns = ['Neighborhood', 'Water Features']
WF_count = WF_count.sort_values(by='Water Features', ascending=False)
print(WF_count)
```

## We see that Squirell Hill South has the most Water Features. 
## Let's bring in a map of the Pittsburgh Neighborhoods, and then plot the data.

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") 
neighborhoods.plot()
```

# The map is in, so now it is time to merge the data, 

```python
WF_map = neighborhoods.merge(WF_count, how='left', left_on='hood', right_on='Neighborhood')
WF_map[['hood','Water Features','geometry']].head()
```

## Time to plot the data on the map. We will use a blue coloured scale. 

```python
WF_map.plot(column='Water Features', 
               cmap='Blues',              
               edgecolor="black",       
               legend=True,              
               legend_kwds={'label': "Number of Water Pokemon"}, 
               figsize=(15, 10),         
               missing_kwds={"color": "lightgrey"} 
               )
```

# Now with Pittsburgh having many rivers, we are also likely to find Water pokemon that border the rivers
# We will add +20 water features for the neighborhoods that border the rivers. The neighborhoods had to be manually looked up for this step 

```python
neighborhoods_to_update = [ "Lincoln-Lemington-Belmar", "Highland Park", "Morningside", "Upper Lawrenceville",
"Central Lawrenceville", "Lower Lawrenceville", "Herrs Island", "Strip District",
"Central Business District", "Troy Hill", "North Shore", "Chateau", "Marshall-Shadeland","Brighton Height",
"Brunot Island", "Esplen", "Elliott", "South Shore", "South Side Flats",
"Hays", "Bluff", "South Oakland", "Hazelwood", "Squirrel Hill South", "Glen Hazel", "Swisshelm Park"]
```

# We can use a for loop to add 20 water features. For Neighborhoods that might not have any water features, 

```python
for hood in neighborhoods_to_update:
    if hood in WF_count['Neighborhood'].values:
        WF_count.loc[WF_count['Neighborhood'] == hood, 'Water Features'] += 20
    else:
        WF_count = pd.concat([WF_count, pd.DataFrame({'Neighborhood': [hood], 'Water Features': [20]})])
```

# And now we can sort and display the updated water features when accounting for river bordering neighborhoods

```python
WF_count = WF_count.sort_values(by='Water Features', ascending=False).reset_index(drop=True)

print(WF_count)
```

# Squirrel Hill South is still where we are going to most likely find water pokemon, but our 3rd, 4th, and 5th neighborhoods have changed.  

# Let's merge the new updated water count

```python
WF_map = neighborhoods.merge(WF_count, how='left', left_on='hood', right_on='Neighborhood')
WF_map[['hood','Water Features','geometry']].head()
```

# And now time to update our map

```python
WF_map.plot(column='Water Features', 
               cmap='Blues',              
               edgecolor="black",       
               legend=True,              
               legend_kwds={'label': "Number of Water Pokemon"}, 
               figsize=(15, 10),         
               missing_kwds={"color": "lightgrey"} 
               )
```
