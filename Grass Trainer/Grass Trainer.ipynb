# Grass Pokemon:
## Grass Pokemon are found all over the world, but they are found in greater numbers in parks. 
## To start off we are going to import the nessesary modules. 

```python
import pandas as pd
import geopandas
import matplotlib.pyplot as plt
%matplotlib inline
```

## Importing the data to be used. 

```python
parks_df = pd.read_csv("City Parks.csv")
```

## Next, a new DataFrame is created called "parks_count". This will present the data into two colums called "Neighborhood" and "Number of Parks"

```python
parks_count = parks_df.groupby("neighborhood")['name'].count().reset_index()
parks_count.columns = ['Neighborhood', 'Number of Parks']
parks_count = parks_count.sort_values(by='Number of Parks', ascending=False)
print(parks_count)
```

# So we now know that "East Liberty" has the highest number of parks per neighborhood.
## A Pokemon trainer is constantly on the move though, and a map is the most basic necessity of any Pokemon Master. Let's go ahead and see where this grass type Pokemon paradise is. 

# Time to import the map of the Greatest City on Earth

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp") 
neighborhoods.plot()
```

# Beautiful map, but it does not tell us where we are gonna find grass pokemon at. 
## Next step is to merge the data into the map. Then to check that the merge happened. 

```python
parks_map = neighborhoods.merge(parks_count, how='left', left_on='hood', right_on='Neighborhood')
parks_map[['hood','Number of Parks','geometry']].head()
```

# Plot the data on the map now

```python
parks_map.plot(column='Number of Parks')
```

## Map does not have a legend and also does not shade to corelate any data. Lets change that. 

```python
parks_map.plot(column='Number of Parks', 
               cmap='Greens',              
               edgecolor="black",       
               legend=True,              
               legend_kwds={'label': "Number of Grass Pokemon"}, 
               figsize=(15, 10),         
               missing_kwds={"color": "lightgrey"} 
               )
```
